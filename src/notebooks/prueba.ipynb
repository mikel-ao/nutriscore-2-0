```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# 1. Cargar el dataset de alimentos procesado (el que unifica 800k registros)
df_final = pd.read_csv('../data/alimentos_con_semaforo_final.csv')
```

```python
sns.pairplot(df_final)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Calculamos la matriz de correlación de Spearman
# Solo sobre las columnas numéricas para evitar errores
corr_matrix = df_final.select_dtypes(include=[np.number]).corr(method='spearman')

# 2. Configuramos el tamaño de la figura
plt.figure(figsize=(12, 10))

# 3. Dibujamos el Heatmap
sns.heatmap(corr_matrix, 
            annot=True,          # Muestra los números dentro de los cuadros
            fmt=".2f",           # Formato con 2 decimales
            cmap='coolwarm',     # Color: azul (negativo), rojo (positivo)
            linewidths=0.5,      # Espacio entre cuadritos
            center=0)            # El color neutro será el 0

plt.title('Heatmap de Correlación de Spearman (Variables Alimentarias)')
plt.show()
```

```python
# 1. Calculamos la matriz de correlación de Pearson (es el método por defecto)
corr_pearson = df_final.select_dtypes(include=[np.number]).corr(method='pearson')

# 2. Configuramos el gráfico
plt.figure(figsize=(12, 10))

# 3. Dibujamos el Heatmap
sns.heatmap(corr_pearson, 
            annot=True, 
            fmt=".2f", 
            cmap='YlGnBu',  # Un mapa de colores distinto para diferenciarlo (Amarillo-Verde-Azul)
            linewidths=0.5,
            center=0)

plt.title('Heatmap de Correlación de Pearson (Relación Lineal)')
plt.show()
```

```python
plt.figure(figsize=(12, 10))

# Dibujamos el heatmap
sns.heatmap(corr_pearson, 
            annot=True, 
            fmt=".2f", 
            cmap='YlGnBu', 
            center=0)

# Rotación de los títulos:
# 'rotation=45' inclina el texto
# 'ha="right"' alinea el final del texto con la marca del eje para que se lea mejor
plt.xticks(rotation=45, ha='right')

plt.title('Heatmap de Correlación de Pearson')
plt.show()
```
